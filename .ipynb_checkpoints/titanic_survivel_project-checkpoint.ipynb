### Convertendo colunas com atributos categóricos

```python
titanic_data.replace({'Sex':{'male': 0, 'female': 1}, 'Embarked':{'S': 0, 'C': 1, 'Q': 2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
#Separando features e o target
X = titanic_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Survived'], axis=1)
Y = titanic_data['Survived']
```

```python
#scaler data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Treinamento do modelo

```python
# Declaração de variáveis

list_activation_func = ['relu', 'tanh', 'logistic', 'identity']
list_learning_rate = ['constant', 'invscaling', 'adaptive']
list_of_solvers = ['lbfgs', 'sgd', 'adam']

number_of_nodes = 256 # Potência de 2

results = []
```

```python
# Treinamento de um conjunto de redes

# Uma camada intermediária
for n in range(1, (number_of_nodes + 1)):
    for act in list_activation_func:
        for lr in list_learning_rate:
            for solver in list_of_solvers:
                clf = MLPClassifier(hidden_layer_sizes=(n,), activation=act, learning_rate=lr, solver=solver)
                clf.fit(X_train, Y_train)
                y_pred = clf.predict(X_test)
                accuracy = accuracy_score(Y_test, y_pred)
                precision = precision_score(Y_test, y_pred, average='weighted')
                recall = recall_score(Y_test, y_pred, average='weighted')
                results.append({
                    'hidden_layer_sizes': (n,),
                    'activation': act,
                    'learning_rate': lr,
                    'solver': solver,
                    'accuracy': accuracy,
                    'precision': precision,
                    'recall': recall
                })

# Duas camadas intermediárias
for n in range(1, (number_of_nodes + 1)):
    for m in range(1, (number_of_nodes + 1)):
        for act in list_activation_func:
            for lr in list_learning_rate:
                for solver in list_of_solvers:
                    clf = MLPClassifier(hidden_layer_sizes=(n, m), activation=act, learning_rate=lr, solver=solver)
                    clf.fit(X_train, Y_train)
                    y_pred = clf.predict(X_test)
                    accuracy = accuracy_score(Y_test, y_pred)
                    precision = precision_score(Y_test, y_pred, average='weighted')
                    recall = recall_score(Y_test, y_pred, average='weighted')
                    results.append({
                        'hidden_layer_sizes': (n, m),
                        'activation': act,
                        'learning_rate': lr,
                        'solver': solver,
                        'accuracy': accuracy,
                        'precision': precision,
                        'recall': recall
                    })
```

```python
# Resultados dos melhores parâmetros

results_df = pd.DataFrame(results)
best_results = results_df.sort_values(by=['accuracy', 'precision', 'recall'], ascending=False).head()

print(best_results)
```
